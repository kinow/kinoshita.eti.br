```python
import pandas as pd
import yaml
```

```python
with open('../../_data/short-stories.yml') as in_file:
    df = pd.io.json.json_normalize(yaml.load(in_file))    
```

```python
pd.set_option('display.max_rows', None)

df\
    .groupby('author')\
    .agg(count=('author', 'count'))\
    .sort_values('count', ascending=False)
```

```python
df\
    .groupby('author')\
    .agg(count=('author', 'count'))\
    .sort_values('author', ascending=True)
```
